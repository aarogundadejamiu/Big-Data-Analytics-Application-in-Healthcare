```python
import pandas as pd

# Read the dataframes
df_1 = pd.read_csv('HCAHPS-Hospital.csv', low_memory=False)
df_2 = pd.read_csv('Hospital_General_Information.csv', low_memory=False)

# Define the columns to check for equality
key_columns = ['Facility ID', 'Address', 'ZIP Code']

for col in key_columns:
    df_1[col] = df_1[col].astype(str).str.strip().str.upper()
    df_2[col] = df_2[col].astype(str).str.strip().str.upper()

# Loop through each row in df_2
for index, row in df_2.iterrows():
    # Find matching rows in df_1 based on key columns
    matching_indices = df_1.index[(df_1[key_columns].values == row[key_columns].values).all(axis=1)]
    
    # If there are matching rows in df_1, merge columns from df_2 to df_1
    if len(matching_indices) > 0:
        for col in df_2.columns[7:]:  # Select columns 8 to last in df_2
            df_1.loc[matching_indices, col] = row[col]
    else:
        # If no matching rows in df_1, skip merging
        pass

# Save the merged dataframe
df_1.to_csv('merged_hospital_data.csv', index=False)
print(df_1)
```

### State Full Name

```python
import us


# Define the conversion function
def convert_abbr_to_full(state_abbr):
    if state_abbr:
        full_state = us.states.lookup(state_abbr)
        return full_state.name if full_state else "Unknown"
    else:
        return "Unknown"

# Apply the full name conversation to the State column and store in a new column
df['Full State Name'] = df['State'].apply(convert_abbr_to_full)

# Get the index of the 'State' column
state_column = 'State'
full_state_column = df.pop('Full State Name')  # Remove the new column from DataFrame and get its values
index_after = df.columns.get_loc(state_column) + 1  # Get the index after the specified column
df.insert(index_after, 'Full State Name', full_state_column)  

# Print the DataFrame with the full state names
print(df[['State', 'Full State Name']])
```

### Dropping rows in Answer Description where Patient Survey Rating is Not Applicable

```python
# Define the list of words to check for in 'HCAHPS Answer Description'
words_to_check = ['rating', 'score', 'star', 'linear', 'mean']

# Create a boolean mask to identify rows to drop
mask = (df['Patient Survey Star Rating'] == 'Not Applicable') & (df['HCAHPS Answer Description'].str.contains('|'.join(words_to_check)))

# Drop rows based on the mask
df = df[~mask].copy()

# Reset inde
df.reset_index(drop=True, inplace=True)

# Now, rows where 'Patient Survey Star Rating' is 'Not Applicable' and 'HCAHPS Answer Description' contains specified words are dropped
filtered_df = df[df['Patient Survey Star Rating'] == 'Not Applicable']

filtered_df[['HCAHPS Answer Description','Patient Survey Star Rating']]
```

### Dropping rows in Answer Description where Patient Survey Rating is Not Available

```python
# Define the list of words to check for in 'HCAHPS Answer Description'
words_to_check = ['rating', 'star']

# Create a boolean mask to identify rows to drop
mask = (df['Patient Survey Star Rating'] == 'Not Available') & (df['HCAHPS Answer Description'].str.contains('|'.join(words_to_check)))

# Drop rows based on the mask
df = df[~mask].copy()

# Reset index if needed
df.reset_index(drop=True, inplace=True)

# Now, rows where 'Patient Survey Star Rating' is 'Not Available' and 'HCAHPS Answer Description' contains specified words are dropped
filtered_df = df[df['Patient Survey Star Rating'] == 'Not Available']

filtered_df[['HCAHPS Answer Description','Patient Survey Star Rating']]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Create Sentiment dataframe excluding HCAHPS Answer Description rows that contain 'star' and 'rating'
df_sentiment = df[['HCAHPS Answer Description']].copy()

# Create a boolean mask to identify rows to keep
words_not_to_select = ['star', 'rating']
mask = ~(df_sentiment['HCAHPS Answer Description'].str.contains('|'.join(words_not_to_select)))

# Apply the boolean mask to filter the DataFrame
df_sentiment = df_sentiment[mask].copy()

# Reset index if needed
df_sentiment.reset_index(drop=True, inplace=True)

# Instantiate the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Apply VADER sentiment analysis to the 'HCAHPS Answer Description' column
df_sentiment['Sentiment Score'] = df_sentiment['HCAHPS Answer Description'].apply(lambda x: sid.polarity_scores(x.lower())['compound'])

# Get unique sentiment scores
unique_sentiment_scores = df_sentiment['Sentiment Score'].unique()

# Create the bin edges
bin_edges = pd.qcut(unique_sentiment_scores, q=5, duplicates='drop', retbins=True)[1]

# Define bin labels
bin_labels = ['Very Negative', 'Negative', 'Neutral', 'Positive', 'Very Positive']

# Create the 'Sentiment Categories' column
df_sentiment['Sentiment Categories'] = pd.cut(df_sentiment['Sentiment Score'], bins=bin_edges, labels=bin_labels, include_lowest=True)

# Print the sentiment categories
print(df_sentiment['Sentiment Categories'].value_counts())
```
